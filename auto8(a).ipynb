```python
#Classical Machine learning
#Regression
import pandas as pd
df = pd.DataFrame({'A':[3,5,7,9],'B':[4,6,8,10],'target':[7,11,15,19]})
df
```

```python
x = df[['A','B']]
y = df['target']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
print(model.coef_)
print(round(model.intercept_,0))
```

```python
model.predict([[10,5]])
```

```python
#STEP 1 (Import all libraries)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#Step2 (read Data)
df = pd.read_csv('/content/diabetes.csv')
df.head()
```

```python
df.columns
```

```python
#step 3 define X and Y
X = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin']]
y = df['BMI']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=32)
```

```python
#STEP 4 (create a pbject of algorithm)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
#STEP 5 fit the model
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
df1 = pd.DataFrame({'Actual_values':y_test,'Predicted_values':y_pred})
df1
```

```python
lr.score(X_train,y_train) * 100
```
